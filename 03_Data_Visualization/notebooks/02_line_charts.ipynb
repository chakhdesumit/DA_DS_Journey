# Line Charts

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
## Path of th file to read
spotify_filepath = "../../datasets/Data_Visualization/spotify.csv"

## Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col = "Date", parse_dates=True)
```

Examin the data :

```python
## Print the first 5 rows of the data
spotify_data.head()
```

```python
## Prin the last five rows of the data
spotify_data.tail()
```

Plot the data :

```python
## Line chart showing daily global streams of each song
sns.lineplot(data = spotify_data)
```

```python
## Ste the width and height of the figure
plt.figure(figsize = (14,6)) ## (figsixe = width, height)

## Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

## Line chart showingh daily global streams of each song
sns.lineplot(data=spotify_data)
```

```python
## Set the width and the height o fthe figures
plt.figure(figsize=(14,6))

## Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

## Line chart showing daily global streams of "Shape of You"
sns.lineplot(data = spotify_data['Shape of You'], label = "Shape of you")

## Line chary showing daily global streams of 'Despacito
sns.lineplot(data = spotify_data['Despacito'], label = "Despacito")

## Add label for horizontal axis
plt.xlabel("Date")
```
