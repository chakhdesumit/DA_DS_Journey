# Lists

```python
myList = ["apple", "banana", "cherry", "kiwi"]
print(myList)
```

```python
## List Length
myList = ["apple", "banana", "cherry", "kiwi"]
print(len(myList))
```

```python
## List items - Data Types
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]

print(type(list1))
print(type(list2))
print(type(list3))
```

```python
## List construtor list()
thislist = list(("apple", "banana", "cherry"))
print(thislist)
```

```python
## Access items
thislist = ["apple", "banana", "cherry"]
print(thislist[2])
```

```python
## Negative indexing
thislist = ["apple", "banana", "cherry"]
print(thislist[-2])
```

```python
## Range of indexes
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
print(thislist[1:3])
print(thislist[5:6])
print(thislist[:3])
print(thislist[2:])
print(thislist[-4:-1])
```

```python
## Check if item exists
thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
    print("Yes, 'apple' is in the fruits lits")
```

```python
## Change items value
thislist = ["apple", "banana", "cherry"]
thislist[1] = "kiwi"
print(thislist)
```

```python
## change a Range of item values
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackberry", "watermelon"]
print(thislist)
```

```python
## cahnge the secondvalue by replacing it with two new values
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:2] = ["blackberry", "watermelon"]
print(thislist)
```

```python
## change the second and third value by replacing it with one value
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["watermelon"]
print(thislist)
```

```python
## insert() : The insert() method inserts an item at the specified index:
thislist = ["apple", "banana", "cherry"]
thislist.insert(2, "watermelon")
print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"]
print(thislist.insert(2, "Mango"))
```

```python
thislist = [3, 1, 2]

print(thislist.sort())
print(thislist)
```

```python
## Append items : To add item to the list..
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

```python
## Extend List : To append elements from another list to the current list, use the extend() method.
thislist = ["apple", "banana", "cherry"]
tropical = ["mango", "pineapple", "papaya"]
thislist.extend(tropical)
print(thislist)
```

```python
## Add any iterable : The extend() method does not have to append lists, you can add any iterable object (tuples, sets, dictionories etc.)
thislist = ["apple", "banana", "cherry"]
thistuple = ("kiwi", "orange")
thislist.extend(thistuple)
print(thislist)
```

```python
## Remove list items : remove()
thislist = ["apple", "banana", "cherry"]
thislist.remove("banana")
print(thislist)
```

```python
## Remove Specified Index : pop() for removing specified index
thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(thislist)
```

```python
## if not specified index the pop() removes the last item
thislist = ["apple", "banana", "cherry"]
thislist.pop()
print(thislist)
```

```python
## The del keyword also removes the specified index
thislist = ["apple", "banana", "cherry"]
del thislist[1]
print(thislist)
```

```python
## if not specified index it can delete whole list
thislist = ["apple", "banana", "cherry"]
del thislist
print(thislist)
```

```python
## clear the list : clear()
thislist = ["apple", "banana", "cherry"]
thislist.clear()
print(thislist)
```

```python
## Loop List
thislist = ["apple", "banana", "cherry"]
for x in thislist:
    print(x)
```

```python
## Loop through the index numbers
thislist = ["apple", "banana", "cherry"]
for i in range(len(thislist)):
    print(thislist[i])
```

```python
## Sort list :sort()
## sort the list alphabatically :
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort()
print(thislist)
```

```python
## sort the list numerically
thislist = [100, 50, 65, 82, 23]
thislist.sort()
print(thislist)
```

```python
## Sort Descending : To sort descending, use the keyword argument reverse = True :
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort(reverse = True)
print(thislist)
```

```python
thislist = [100, 50, 65, 82, 23]
thislist.sort(reverse = True)
print(thislist)
```

```python
## reverse Order: the reverse() method reverse the current sorting order of the elements
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

```python
## Copy a List : 
thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
```

```python
thislist = ["apple", "banana", "cherry"]
mylist = thislist
print(mylist)
```

```python
## list() :
thislist = ["apple", "banana", "cherry"]
mylist = list(thislist)
print(mylist)
```

```python
## Slice Operator [:] :-
thislist = ["apple", "banana", "cherry"]
mylist = thislist[:]
print(mylist)
```

```python
## Join Lists
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 +list2
print(list3)
```
