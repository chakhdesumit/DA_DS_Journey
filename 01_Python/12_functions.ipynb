# Functions

A function is a block of code which only runs when it is called.

A function can return data as a result.

A function helps avoiding code repetition.

```python
## A function is defined using the def keyword, followed by a function name and parentheses:
def my_function():
    print("Hello from a function")
```

```python
## Calling a Function : To call a function, write its name followed by parantheses:
def my_function():
    print("Hello from a function")

my_function()
```

```python
def my_funcion():
    print("Hello from a function")

my_function()
my_function()
my_function()
```

### Why Use Functions?

Imagine you need to convert temperatures from Fahrenheit to Celsius several times in your program. Without functions, you would have to write the same calculation code repeatedly:

```python
temp1 = 77
celsius1 = (temp1 - 32) * 5 / 9
print(celsius1)

temp2 = 95
celsius2 = (temp2 - 32) * 5 / 9
print(celsius2)

temp3 = 50
celsius3 = (temp3 - 32) * 5 / 9
print(celsius3)
```

```python
## With functions, you write the code once and reuse it:
def fahrenheit_to_celsius(fahrenheit):
    return(fahrenheit - 32) * 5 / 9

print(fahrenheit_to_celsius(77))
print(fahrenheit_to_celsius(95))
print(fahrenheit_to_celsius(50))
```

```python
## Return Values : Functions can send data back to the code that called them using the return statement. When a function reaches a return statement, it stops executing and sends the result back :

def get_greetings():
    return "Hello from a function"

message = get_greetings()
print(message)
```

```python
## We can use the returned value directly:
def get_greetings():
    return "Hello from a function"

print(get_greetings())
```

```python
## Arguments : Information can be passed into functions as arguments. Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.

def my_function(fname):
    print(fname + " Refsnes")
    
my_function("Emil")
my_function("Tobias")
my_function("Linus")
```

```python
## Paramenters vs Arguments : 
## A parameter is the variable listed inside the parentheses in the function definition.
## An argument is the actual value that is sent to the function when it is called.

def my_function(name):      ## name is a parameter
    print("Hello", name)

my_function("Sumit")        ## Sumit is an argument
```

```python
## Number of Arguments :
def my_function(fname, lname):
    print(fname + " " + lname)

my_function("Sumit", "Chakhde")
```

```python
## Default Parameter Values: We can assign default values to parameteres. if the function is caleld without and argument, it uses the default value:

def my_function(name = "friend"):
    print("Hello", name)

my_function("Sumit")
my_function("Chakhde")
my_function()
my_function("George")
```

```python
def my_function(country = "Norway"):
    print("I am from", country)

my_function("Sweden")
my_function("India")
my_function()
my_function("Brazil")
```

```python
## Keyword Arguments : we can send arguments witht he key = value syntax:

def my_function(animal, name):
    print("I have a", animal)
    print("My", animal + "'s name is", name)

my_function(animal = "dog", name = "Shiro")
```

```python
## This way, with keword arguments, the order of the arguments doesn't matter :
def my_function(animal, name):
    print("I have a", animal)
    print("My", animal + "'s name is", name)

my_function( name = "Shiro", animal = "dog")
```

```python
## Possitional Arguments :
## When we call a function with arguments without using keywords, they are called positional arguments. Positional arguments must be in the correct order:

def my_function(animal, name):
    print("I have a", animal)
    print("My", animal + "'s name is", name)

my_function("dog", "Shiro")
    
```

```python
## Return Values : Functions can return values using the return statement:
def my_function(x, y):
    return x + y

result = my_function(5,3)
print(result)
```

```python
def my_function():
  return ["apple", "banana", "cherry"]

fruits = my_function()
print(fruits[0])
print(fruits[1])
print(fruits[2])
```

### *args and ** Kwargs

By default, a function must be called with the correct number of arguments.

However, sometimes you may not know how many arguments that will be passed into your function.

*args and **kwargs allow functions to accept a unknown number of arguments.

```python
## Arbitrary Arguments - *args
def my_function(*kids):
    print("The youngest child is " + kids[2])
    
my_function("Emil", "Tobias", "Linus")
```

```python
## *args : 
def my_function(*args):
    print("Type : ", type(args))
    print("First argument : ", args[0])
    print("Second argument : ", args[1])
    print("All arguments : ", args)
    
my_function("Emil", "Tobias", "Linus")
```

```python
## Using *args with Regular Arguments :
def my_function(greeting, *name):
    for name in name:
        print(greeting, name)
        
my_function("Hello", "Emil", "Tobias", "Linus")
```

```python
## A function that calculates the sum of any number of values :
def my_function(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(my_function(1, 2, 3))
print(my_function(10, 20, 30, 40))
print(my_function(5))
```

```python
## finding the maximym value:
def my_function(*numbers):
    if len(numbers) == 0:
        return None
    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
        return max_num

print(my_function(3, 7, 2, 9, 1))
```

Lambda Functions :

A lambda function is a small anonymous function

A lambda function can take any numver of argments, but can only have one expression

Syntax :

lambda arguments : expression

```python
x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a, b : a * b
print(x(6, 9))
```

```python
a = lambda x, y, z : x + y + z
print(a(5, 6, 2))
```

```python
def myfunc(n) :
    return lambda a : a * n

mydoubler  = myfunc(2)

print(mydoubler(11))
```
