# Python Variables

### Creating variables

```python
x = 5
y = "Sumit"

print(x,y)
```

Type of Variable

```python
x = 5
y = "Sumit"

print(type(x))
print(type(y))
```

Single and Double Quotes

```python
x = "Sumit"
y = 'Sumit'

print(x)
print(y)
```

Case Sensitive

```python
A= 4
A = "Sally"
## A = 4 will overwrite aA = "Sally"
print(A)
```

```python
a = 4
A = "Sally"
## A will not overwrite a
print(a)
print(A)
```

Many values to multiple variables

```python
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
```

Ome value to multiple variables

```python
x = y = z = "Orange"
print(x)
print(y)
print(z)
```

### Unpack a Collection
If you have a collection of values in a list, tuple etc. Python allows you to extract the values into variables. This is called unpacking.

```python
fruits = ["apple", "banana", "cherry"]
x, y, z = fruits
print(x)
print(y)
print(z)
```
