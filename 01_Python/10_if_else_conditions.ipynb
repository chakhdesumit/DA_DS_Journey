# If Statements

```python
## if statement : if keyword
a = 33
b = 200
if b > a:
    print(f"{b} is greater then {a}")
```

```python
## multiple statement in if block
age = 20
if age >= 18:
    print("You are and adult")
    print("You can vote")
    print("You have full legal rights")
```

```python
age = int(input("Enter your age : "))
if age >= 18:
    print("You are an adult")
else:
    print("you are minor")
```

```python
try:
    age = int(input("Enter your age : "))
    if age >= 18:
        print("You are an adult")
    else:
        print("you are minor")
except ValueError:
    print("Please enter a valid number")
```

```python
#3 The Elif Keyword: 
a = 33
b = 33
if b> a:
    print("b is greater than a")
elif a ==b:
    print("a and b are equal")
```

```python
## Multiple elif statement :
score = int(input("Enter the score : "))

if score >= 90:
    print("Grade : A")
elif score >= 80:
    print("Grade : B")
elif score >= 70:
    print("Grade : C")
elif score >= 60:
    print("Grade : D")
else :
    print("Grade : F")
```

```python
## Short Hand If: one line if statement
a = 5
b = 2
if a > b: print("a is greater than b")
```

```python
a = 2
b = 330
print(f"{a}") if a > b else print(f"{b}")
```

```python
## Nested if statements :
x = 41

if x > 10:
    print("Above ten,")
    if x > 20:
        print("and also above 20!")
    else:
        print("but not above 20")
```

```python
age = int(input("Enter your age : "))
has_license =  True

if age >= 18:
    if has_license :
        print("You can drive")
    else:
        print("You need a license")
else:
    print("You are too young to drive")
```

```python
age = int(input("Enter your age : "))

if age >= 18:
    license_input = input("Do you have a license? (yes/no) : ").lower() ## yes, Yes, YES
    
    if license_input == "yes":
        print("You can drive")
    else:
        print("You need a license")
else:
    print("You are too young to drive")
```
